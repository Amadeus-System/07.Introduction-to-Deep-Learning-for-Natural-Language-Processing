## 3) 간단한 seq2seq 만들기(Simple seq2seq)

앞서 구현한 seq2seq 모델은 지금까지 만든 모델 중에서 가장 복잡한 모델이었다. 아직은 함수형 API의 사용이 부담스러운 사람들을 위해 여기서는 좀 더 단순하고 구현이 간단한 seq2seq를 만들면서 seq2seq의 아키텍처 개념을 이해해보겠다.


### 1.간단한 Seq2Seq(Simple Sequence-to-Sequence)

이번 seq2seq는 보다 단순한 모델이지만 seq2seq의 아이디어는 그대로이다. 인코더는 입력 시퀀스로부터 정보를 압축한 벡터를 생성하고, 디코더는 이 벡터를 전달받아 출력 시퀀스를 만들어낸다. 그리고 입력 시퀀스와 출력 시퀀스의 길이는 다를 수 있다.

<img src = 'https://wikidocs.net/images/page/43646/simpleseq2seq1.PNG' width = 60%>

위의 그림은 인코더가 길이 n인 입력 시퀀스를 입력받아 어떤 의문의 벡터 ?에 입력 정보를 압축하고, 디코더가 이 벡터를 전달받아 길이 m의 출력 시퀀스를 만드는 것을 보여준다. 그렇다면 여기서 인코더가 디코더에게 전달해주는 벡터의 정체는 무엇일까?

<img src = 'https://wikidocs.net/images/page/43646/simpleseq2seq2.PNG' width = 60%>

이번에도 인코더가 디코더에게 전달할 벡터는 인코더 LSTM 셀의 마지막 시점(time step)의 은닉 상태(hidden state)이다. n번째 시점의 은닉 상태는 입력 시퀀스의 정보들을 반영하고 있다. 앞서 이 원리를 사용하여 텍스트 분류 문제를 풀었었고, 앞서 배운 seq2seq에서도 이를 디코더에게 전달하는 벡터로 사용했다. 그런데 이번에는 디코더로의 전달 방식이 다르다.

<img src = 'https://wikidocs.net/images/page/43646/simpleseq2seq3.PNG' width = 60%>

이번에는 인코더의 마지막 은닉 상태를 디코더의 모든 시점의 입력으로 사용한다. 그리고 이를 위해 입력을 원하는만큼 반복해서 사용하게 해주는 도구인 RepeatVector()를 사용한다. 이게 이 구조의 전부이다. 위의 그림은 RepeatVector()를 사용해서 만든 seq2seq를 보여준다. 구조가 간단한 만큼 코드도 굉장히 간단해진다. 코드의 형식은 아래와 같다.

``` python
model = Sequential()
model.add(LSTM(..., input_shape=(...))) # 인코더
model.add(RepeatVector(...))
model.add(LSTM(..., return_sequences=True)) # 디코더
model.add(TimeDistributed(Dense(...)))
```

앞서 배운 seq2seq 코드에 비하면 굉장히 간단해졌다. 이 구조를 사용한 실습은 향후 추가 예정이다.
